````python
import requests
from bs4 import BeautifulSoup
````

````python
url = 'http://www.mineral.k12.nv.us/pages/School_Board_Minutes'
response = requests.get(url, verify=False)
doc = BeautifulSoup(response.text, 'html.parser')
````

````python
doc.text
print(doc.prettify())
````

````python
doc.text
````

````python
# die urls
baseurl="http://www.mineral.k12.nv.us"
urls = doc.find_all('a',href=True)

url_list = []
for url in urls: 
    url_list.append(url['href'])
    print(url['href'])
````

````python
urls
````

````python
dates = doc.find_all(style="color: #000000;")

dates_list =[]
for date in dates:
    dates_list.append(date.text)
    print(date.text)
````

````python
import pandas as pd

schoollist = pd.DataFrame(list(zip(dates_list,url_list)),columns=['dates','url'])
````

````python
baseurl="http://www.mineral.k12.nv.us"
schoollist2 = schoollist[5:]
schoollist2['url'] = baseurl + schoollist2['url']
````

````python
import datetime

schoollist_def = schoollist2[0:80]

schoollist_def
#schoollist_def['dates'] = pd.to_datetime(schoollist_def['dates'])
````

````python
schoollist_def[schoollist_def['dates'].str.contains(r'\d\d\d\d')]
````

````python
sl = schoollist_def[schoollist_def['dates'].str.contains(r'\d\d\d\d')]


sl['dates'] = pd.to_datetime(sl['dates'], errors='coerce')

sl
````

````python
sl.to_csv("minutes.csv", index=False)
````
