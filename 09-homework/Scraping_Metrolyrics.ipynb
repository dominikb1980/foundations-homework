```python
df = pd.read_csv("metrolyrics.csv")
df.head()
```

```python
# eine Funktion bauen, def => diese dann anwenden

def scrape_page(row):
    url = f"{row['internet']}"
    # url = f"http://www.metrolyrics.com/selingkar-kasih-lyrics-rem.html"
    print("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    rows = []
    
    lyrics = doc.find_all(class_="verse")
    
    for lyric in lyrics:
        
        row = {}
       
        row['lyrics'] = lyric.text.strip()
        rows.append(row)
        
        return pd.Series(row)
        
df.apply(scrape_page, axis=1)
```

```python
scraped_df = df.apply(scrape_page, axis=1)
```

```python
scraped_df.head(20)
```

```python
df.merge(scraped_df, left_index=True, right_index=True)
```

```python
df.to_csv("metrolyrics2.csv", index=False)
```
