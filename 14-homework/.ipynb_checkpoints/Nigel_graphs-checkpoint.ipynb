```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv("data/scrabble-point-spread.csv")
df.head()
```

```python
df1 = pd.read_csv("data/scrabble-tournament.csv")
df1.head()

df1['player_id'] = df1['Player ID']

df1.head()
```

```python
df3 = df.merge(df1, left_on='player_id', right_on='player_id')
df3.head()
```

```python
def build_colors(row):
    if row['division'] == 'Divison 1':
        return 'red'
    elif row['division'] == 'Division 2':
        return 'yellow'
    elif row['division'] == 'Division 3':
        return 'green'
    else:
        return 'lightblue'

colors = df.reset_index().apply(build_colors, axis=1)
And then you use it by passing color= to your .plot method:

df3.plot(x='points_against', y='points_for', kind='scatter', color=colors)
```
